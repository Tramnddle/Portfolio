## Change date type of DOB

```python
df_demo.DOB=pd.to_datetime(df_demo.DOB, format='%Y-%m-%d')
```

```python
from datetime import date

def calculate_age(born):
    today = date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))
```

```python
df_demo.DOB=df_demo.DOB.apply(calculate_age)
```
