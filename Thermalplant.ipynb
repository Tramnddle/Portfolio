```python
# import thu vien de stack
import matplotlib.patches as mpatches

thermal = test[test.Technology == 'Thermal'][['Owner Name', 'C']] # Lay rieng thermal
wind = test[test.Technology == 'Wind'][['Owner Name', 'C']] # Lay rieng wind

plt.figure(figsize=(10,10))
# ve bar 1
bar1 = sns.barplot(data = thermal, x = 'Owner Name', y = 'C', color = 'green')

# ve bar 2
bar2 = sns.barplot(data = wind, x = 'Owner Name', y = 'C', color = 'orange')

# stack 2 do thi vua tao va tao legend
top = mpatches.Patch(color='green', label='thermal')
bottom = mpatches.Patch(color='orange', label='wind')
plt.legend(handles = [top,bottom])

plt.show()
```

```python
Active_Capacities = AppendData[AppendData['Status']=="Active"].groupby(['Technology', 'Owner Name'])['Total Capacity (MW)'].sum().to_frame()
```

```python
Active_Capacities_All = AppendData[AppendData['Status']=="Active"].groupby(['Owner Name'])['Total Capacity (MW)'].sum().reset_index()
Active_Capacities_All
```

```python
Active_Capacities_Wind = AppendData[(AppendData['Status']=="Active") & (AppendData['Technology'] == 'Wind')].groupby(['Owner Name'])['Total Capacity (MW)'].sum().reset_index()
Active_Capacities_Wind
```

```python
#Tram chinh figzie cho no to ra nua la ok nhe

import matplotlib.patches as mpatches

plt.figure(figsize=(14,14)) #<- them figsize de chinh kich co
# Stackbar can 1 cai la gia tri tong
bar1 = sns.barplot(data = Active_Capacities_All, x = 'Owner Name', y ='Total Capacity (MW)', color = 'darkblue' ) # Tong cap
bar2 = sns.barplot(data = Active_Capacities_Wind, x = 'Owner Name', y ='Total Capacity (MW)', color = 'lightblue' ) # Wind cap
# Them 1 so customize de hien legend
top = mpatches.Patch(color = 'darkblue', label = 'Thermal')
bottom = mpatches.Patch(color = 'lightblue', label = 'Wind')

plt.legend(handles = [top, bottom])
plt.show()
```
